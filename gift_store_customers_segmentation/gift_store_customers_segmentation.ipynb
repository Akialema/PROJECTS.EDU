<h3 id='refundnum'>Определение количества возвратов</h3>Задание 3.4

>Давайте подробнее разберёмся с возвратами. Сейчас на каждый уникальный товар заведена отдельная строка в таблице — это мешает определить общее количество возвратов.
>
>Чтобы подсчитать число возвратов, сначала необходимо определить, сколько уникальных товаров указано в транзакции (корзине/basket) для каждой уникальной пары «клиент-заказ». 
>
>Теперь мы можем трезво оценить, как часто клиенты возвращают товар.

Добавьте во вспомогательную таблицу `nb_products_per_basket` признак отмены заказа (`order_canceled`): он равен `1`, если транзакция является возвратом, и `0` — в противном случае.

Сколько процентов заказов в таблице являются возвратами? Ответ приведите в виде целого числа.

P.S. В признаке `InvoiceNo` присутствует либо число, либо число с буквой. Эта буква только одна и это `C`. Мой код не учитывает возможность иной буквы. Нужно быть внимательным/ной.

```python
nb_products_per_basket = df.groupby(['CustomerID', 'InvoiceNo'])['InvoiceDate'].count().reset_index(name='Number of products')

# Здесь:
# Из строки сначала выделяется только буква; (.str.extract(r'([A-Za-z]+)'))
# Т.к. создаётся датафрейм - берется единственный столбец с индексом [0];
# Определяется: кто NaN, а кто нет; (.notna()) <-- чтобы не NaN был True
# .astype(int) меняет True на 1, False на 0
nb_products_per_basket['order_canceled'] = nb_products_per_basket['InvoiceNo'].str.extract(r'([A-Za-z]+)')[0].notna().astype(int)
display(nb_products_per_basket.head(3))

good_print('Сколько процентов заказов в таблице являются возвратами?', f"{nb_products_per_basket['order_canceled'].value_counts(normalize=True)[1]*100:.0f}%")
print(f'📌 Интересный факт: если бы мы подсчитали количество транзакций, содержащих признак возврата, в изначальной таблице,\n где на каждый уникальный товар заведена отдельная строка, то количество возвратов было бы менее 1 %. Это некорректное число.')
```

Метод заключается в группировке клиентов на основе следующих параметров:

* `Recency` (Давность) — давность последней покупки клиента;
* `Frequency` (Частота) — общее количество покупок клиента;
* `Monetary Value` (Денежная ценность) — сколько денег потратил клиент.

Суть RFM-анализа состоит в том, что мы разделяем всех клиентов на группы в зависимости от того, как давно они сделали последнюю покупку, как часто покупали и насколько большой была сумма их заказов.
</div>

Например, вот так может выглядеть интерпретация кластеров для случая RF-сегментации (анализа на основе давности и частоты заказов клиента):

<img src='data/pics/rfm-1.png' width='70%' style='display:block; margin:auto;' />

Мы можем рассчитать RFM-характеристики для каждого из клиентов в нашем датасете и на их основе с помощью методов кластеризации построить подобные сегменты клиентов, привязанные к нашим данным.

Чтобы получить RFM-таблицу, нам необходимо сгруппировать данные по идентификаторам клиента и рассчитать следующие агрегированные характеристики:

* `Recency` для $i$-ого клиента рассчитывается как разница между датой и временем последнего заказа и точкой отсчёта, переведённая в дни:  
$t_0-max(t_{i1}, \ t_{i2}, \ ..., \ t_{iM})$, где $t_{ij}$ — дата и время совершения $i$-ым клиентом своей $j$-ой покупки.<br /><br />
В качестве точки отсчёта берём дату на один день «старше» всех дат в наших данных. Это будет 10 декабря 2011 года (в формате datetime — '2011-12-10 00:00:00')<br /><br />
* `Frequency` рассчитывается как общее количество уникальных заказов, которые совершил $i$-ый клиент.<br /><br />
* `Monetary Value` рассчитывается как общая сумма денег, которую $i$-ый клиент потратил на товары (с учётом возвратов).

Создайте `pipeline`, в котором будут следующие шаги:

Мы построили модель кластеризации для части клиентов.Но как сделать предсказание сегмента для клиентов, которые не попали в обучающую выборку?

Вы, вероятно, скажете: «Воспользоваться методом `predict()`!»

Верно, однако всё не так просто.

Вспомним, что мы обучали наши алгоритмы кластеризации на сжатом с помощью `t-SNE` пространстве признаков. То есть, чтобы сделать предсказание для новых объектов, не попавших в выборку, нам необходимо будет совершить трансформацию признаков этих объектов в новое сжатое пространство. 

Однако проблема `t-SNE` заключается в том, что алгоритм непараметрический. Это значит, что он, в отличие от таких алгоритмов, как `PCA` и `SVD`, не создаёт явной функции отображения. Проще говоря, алгоритм трансформирует пространство признаков для обучающей выборки, но не запоминает, каким образом это делает. Это значит, что, передав в алгоритм новые объекты для сжатия пространства, мы получим абсолютно новое представление пространства, не имеющее отношения к тому, что мы построили для обучающей выборки. То есть произойдёт искажение, что может привести к неверным результатам кластеризации.

Именно поэтому у класса `TSNE` из библиотеки `sklearn` нет такого метода, как `transform()`, у него есть только метод `fit_transform()` — заново обучить алгоритм `t-SNE` и произвести трансформацию.

Как же тогда производить кластеризацию для новых объектов, если мы не можем сжимать размерность для новых данных?

Давайте сведём задачу кластеризации к задаче классификации. Действительно, у нас теперь есть истинные метки кластеров, а есть клиенты, которые описываются RFM-характеристиками. Давайте обучим модель, которая на основе RFM-характеристик будет предсказывать клиентский сегмент.

<h3 id='xgbustopt'>Обучение XGBoost через Optuna</h3>

```python
# Создаем DMatrix объекты
dtrain = xgb.DMatrix(X_train_scaled, label=y_train, feature_names=list(rfm_df.columns))
dvalid = xgb.DMatrix(X_test_scaled, label=y_test, feature_names=list(rfm_df.columns))

def objective(trial):
    # Отключаем логи Optuna для этой функции
    optuna.logging.set_verbosity(optuna.logging.WARNING) #По умолчанию: optuna.logging.INFO
    
    # Определяем пространство поиска параметров
    param = {'min_child_weight': trial.suggest_int('min_child_weight', 5, 25),
             'eta': trial.suggest_float('eta', 0.01, 0.20),
             'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 0.9),
             'max_depth': trial.suggest_int('max_depth', 4, 9),
             'subsample': trial.suggest_float('subsample', 0.4, 1.0),
             'lambda': trial.suggest_float('lambda', 0.1, 2.0),
             'nthread': -1,
             'booster': 'gbtree',
             'eval_metric': 'mlogloss',
             'objective': 'multi:softmax',
             'num_class': len(np.unique(y_train))}
    
    # Тренируем модель с текущими параметрами
    evals_result = {}  # Словарь для хранения результатов оценки
    
    model = xgb.train(params=param,
                      dtrain=dtrain,
                      num_boost_round=500,
                      evals=[(dtrain, 'train'), (dvalid, 'valid')],
                      early_stopping_rounds=20,
                      evals_result=evals_result,
                      verbose_eval=False)
    
    # Возвращаем значение метрики для оптимизации
    return evals_result['valid']['mlogloss'][-1]  # Берем последнее значение метрики

# Создаем исследование Optuna
study = optuna.create_study(direction='minimize',
                            pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=5))

# Запускаем оптимизацию
study.optimize(objective, n_trials=50, show_progress_bar=True)

# Выводим лучшие параметры
print("Лучшие параметры:", study.best_params)
print("Лучшее значение:", study.best_value)

# Обучаем финальную модель с лучшими параметрами
best_params = study.best_params
best_params.update({'nthread': -1,
                    'booster': 'gbtree',
                    'eval_metric': 'mlogloss',
                    'objective': 'multi:softmax',
                    'num_class': len(np.unique(y_train))})

# Финальное обучение с лучшими параметрами
final_model = xgb.train(params=best_params,
                        dtrain=dtrain,
                        num_boost_round=500,
                        evals=[(dtrain, 'train'), (dvalid, 'valid')],
                        early_stopping_rounds=20,
                        verbose_eval=100)
```

```python
display(print_metrics(y_train=y_train,
                      y_train_pred=[final_model.predict(dtrain)],
                      y_test=y_test,
                      y_test_pred=[final_model.predict(dvalid)],
                      list_of_metrics=['accuracy_score', 'precision_score', 'recall_score', 'f1_score'],
                      metric_params_dict = {'precision_score': {'average': 'macro'},
                                            'recall_score': {'average': 'macro'},
                                            'f1_score': {'average': 'macro'}}).round(3).T)

ft_importances = pd.DataFrame([final_model.get_score(importance_type='weight')]).T.sort_values(by=0, ascending=False).reset_index()

fig, ax = plt.subplots(figsize=(8, 3))
ax = sns.barplot(y=ft_importances['index'], x=ft_importances[0], palette='deep', hue=ft_importances['index'])
[ax.bar_label(ax.containers[i], fmt='%.0f', label_type='center', color='white', fontsize=8) for i in range(len(ax.containers))]
ax.set(xlabel='P-power!', ylabel='', title='Feature importances for final xgb model')
plt.show('png')
```
