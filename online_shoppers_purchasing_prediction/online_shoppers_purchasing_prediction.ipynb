<h3 id='ex_6_3'>Обучение RandomForest без параметров</h3>

Коллеги посоветовали нам использовать случайный лес (`Random Forest`) для решения данной задачи. Давайте последуем их совету.

Оцените качество такой модели с помощью кросс-валидации по `пяти` фолдам. Так как классы несбалансированы, используйте кросс-валидатор `StratifiedKFold` (перемешивать выборку не нужно).

```python
models_6_5 = [
    RandomForestClassifier(max_depth=5,
                           n_estimators=200,
                           min_samples_leaf=5,
                           random_state=42),
    RandomForestClassifier(max_depth=7,
                           n_estimators=200,
                           min_samples_leaf=5,
                           random_state=42),
    RandomForestClassifier(max_depth=12,
                           n_estimators=200,
                           min_samples_leaf=5,
                           random_state=42)
]
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

for i, j in enumerate(models_6_5):
    
    plot_learning_curve(model=j, X=X_train, y=y_train, # модель, х, у.
                        cv=StratifiedKFold(n_splits=5), # кросс валидатор.
                        scoring='f1', ax=ax[i], # метрика и ах.
                        # тут короче я выдрал из параметров модели max_depth и пиханул в титл.
                        title=' '.join(map(str, pd.Series(j.get_params()).to_frame().reset_index().iloc[4].to_list())))

plt.show()
```
